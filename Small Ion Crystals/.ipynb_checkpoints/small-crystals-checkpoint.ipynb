# Small Ion Crystal Geometries
Ions trapped in a linear Paul trap can be squeezed into 2D crystals. The geometry of the crystal is quick different for small ion numbers. This notebook calculates what those geometries will look like and relative distances between ions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
q = 1.60217662 * 1e-19
ep_0 = 8.85418782 * 1e-12
m = 170.9363258 * 1.66054* 1e-27
w_r = 2*np.pi * 2e6
```

## Confining Potential
The effective confining potential for 2D ion crystals is a harmonic oscillator with frequency $\omega_r$ and radial distance $r$.

$$
V_h = \sum_i^N\frac{1}{2}m \omega_r^2 r_i^2
$$

```python
# x, y grid
x = np.linspace(-1000, 1000, 2000) * 1e-6
y = np.linspace(-1000, 1000, 2000) * 1e-6
XX, YY = np.meshgrid(x, y)

# Compute trapping potential energy
Vh = 1/2 * m * w_r**2 * (XX**2 + YY**2)

# Plot trapping potetnial
plt.figure(figsize=(6,6))
plt.contour(XX, YY, Vh)
plt.xlabel('x (um)')
plt.ylabel('y (um)')
plt.title('Harmonic Trapping Potential')
plt.show()
```

## Ion-ion Interaction Energy
The ion's repel each other through the Coulomb interactions. The total Coulomb energy is the potential at each ion (due to the other ions) times it's charge.

$$
V_c =\frac{q^2}{4\pi \epsilon_0} \sum_i^N\sum_{j\ne i}^N\frac{1}{|r_i - r_j|}
$$

Minimizing these potentials will give the ion equilibrium positions.

```python
# Parameters
n_ions = 2

# Initial position guess
ion_pos = np.array([[3,4], [0, 500]]) * 1e-6

if n_ions != ion_pos.shape[0]:
    print('not all ions have positions')

# Compute trapping potential energy
def compute_Vh(ion_pos):
    Vh_contants = 1/2 * m * w_r**2
    return Vh_contants * np.sum(np.linalg.norm(ion_pos, 2 , axis=1)**2)

# Compute coloumb interaction energy
def compute_Vc(ion_pos):
    Vc_distance = 0
    for x_i, y_i in ion_pos:
        for x_j, y_j in ion_pos:
            if x_i != x_j or y_i != y_j:
                Vc_distance += 1 / np.abs(np.sqrt(x_i**2 + y_i**2) - np.sqrt(x_j**2 + y_j**2))
    return Vc_contants * Vc_distance

Vc = compute_Vc(ion_pos)
Vh = compute_Vh(ion_pos)

Vh, Vc
```
